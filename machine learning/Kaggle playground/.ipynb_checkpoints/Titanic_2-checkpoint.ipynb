```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('titanic/train.csv')
train_data.head()

train_data.fillna(0, inplace=True)
```

```python
feat_names = ['Pclass', 'Age', 'Parch', 'Fare', 'SibSp']
X = train_data.loc[:, feat_names]
y = train_data.Survived

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
score_table = []
highest = (0, 0)
for current_c in range(1,1000):
    c_logreg = LogisticRegression(C=current_c, solver='lbfgs')
    c_logreg.fit(X_train, y_train)
    
    c_score = c_logreg.score(X_test, y_test)
    score_table.append((current_c, c_score))
    
    if c_score > highest[1]:
        highest = (current_c, c_score)
```

```python
highest
```

```python
score_table
```

```python
train_data.head()
```

```python
test_data = pd.read_csv('titanic/test.csv')

X = train_data.loc[:, feat_names]
y = train_data.Survived
X_new = test_data.loc[:, feat_names]

best_logreg = LogisticRegression(C=highest[0])
```

```python
best_logreg.fit(X, y)
```

```python
X_new.fillna(0, inplace=True)
prediction = best_logreg.predict(X_new)
```

```python
pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': prediction}).set_index('PassengerId').to_csv('sub2.csv')
```
